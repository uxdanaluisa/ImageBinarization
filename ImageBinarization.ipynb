```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt

# Abre a janelinha para você escolher a imagem do seu computador
uploaded = files.upload()

# Pega o nome do arquivo que você fez upload
input_image = list(uploaded.keys())[0]
print("Arquivo carregado:", input_image)

# ---- Converter para tons de cinza ----
img = Image.open(input_image).convert("RGB")
gray_img = img.convert("L")
gray_img.save("imagem_gray.png")

# ---- Converter para binária (preto e branco) ----
threshold = 128
binary_img = gray_img.point(lambda p: 255 if p > threshold else 0)
binary_img.save("imagem_binary.png")

# Mostrar resultados
fig, axs = plt.subplots(1, 3, figsize=(12,4))

axs[0].imshow(img)
axs[0].set_title("Original")
axs[0].axis("off")

axs[1].imshow(gray_img, cmap="gray")
axs[1].set_title("Cinza")
axs[1].axis("off")

axs[2].imshow(binary_img, cmap="gray")
axs[2].set_title("Binária")
axs[2].axis("off")

plt.show()
```
